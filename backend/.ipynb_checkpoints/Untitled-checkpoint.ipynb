```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
csv_file = "metrics_with_ablation.csv"
df = pd.read_csv(csv_file)

# Show first 5 rows
df.head()
```

```python
# Pick top 10 videos by first appearance
top_videos = df['video_id'].unique()[:10]
df_top = df[df['video_id'].isin(top_videos)]
# Pick top 10 videos by first appearance
top_videos = df['video_id'].unique()[:10]
df_top = df[df['video_id'].isin(top_videos)]

sns.lineplot(data=df_top, x="window_size", y="MRR@10", hue="video_id", marker="o", linewidth=2)
plt.title("MRR@10 (Top 10 Videos)")
plt.xlabel("Window Size")
plt.ylabel("MRR@10")
plt.legend(title="Video ID", frameon=True)
sns.despine()
plt.show()

sns.lineplot(data=df_top, x="window_size", y="MRR@10", hue="video_id", marker="o", linewidth=2)
plt.title("MRR@10 (Top 10 Videos)")
plt.xlabel("Window Size")
plt.ylabel("MRR@10")
plt.legend(title="Video ID", frameon=True)
sns.despine()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pick top 20 videos by first appearance
top_videos = df['video_id'].unique()[:20]
df_top = df[df['video_id'].isin(top_videos)]

plt.figure(figsize=(14, 7))
sns.set_style("whitegrid")

# Lineplot with markers, solid lines, and better palette
sns.lineplot(
    data=df_top,
    x="window_size",
    y="avg_latency",
    hue="video_id",
    style="video_id",   # different line styles
    markers=True,       # show markers
    dashes=False,       # solid lines
    linewidth=2,
    palette="tab20"     # palette for up to 20 colors
)

# Titles and labels
plt.title("Average Latency per Video and Window Size (Top 20 Videos)", fontsize=16, weight='bold')
plt.xlabel("Window Size", fontsize=12)
plt.ylabel("Average Latency (seconds)", fontsize=12)

# Legend outside plot
plt.legend(title="Video ID", bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

# Remove top and right spines
sns.despine(trim=True)

plt.tight_layout()
plt.show()
```
