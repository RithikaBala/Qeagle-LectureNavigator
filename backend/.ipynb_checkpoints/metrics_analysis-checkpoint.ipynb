```python
# %% [markdown]
# # Lecture Navigator Metrics Analysis
# This notebook visualizes MRR@10, avg latency, and p95 latency per video and window size.

# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# %% [markdown]
# ## Load metrics CSV
# Make sure the CSV was generated by the `metrics_with_ablation.py` script.

# %%
csv_file = "metrics_with_ablation.csv"
df = pd.read_csv(csv_file)
df

# %% [markdown]
# ## MRR@10 Comparison by Window Size

# %%
plt.figure(figsize=(10, 6))
sns.barplot(x="video_id", y="MRR@10", hue="window_size", data=df)
plt.title("MRR@10 by Video and Window Size")
plt.xlabel("Video ID")
plt.ylabel("MRR@10")
plt.xticks(rotation=45)
plt.legend(title="Window Size (s)")
plt.tight_layout()
plt.show()

# %% [markdown]
# ## Average Latency Comparison by Window Size

# %%
plt.figure(figsize=(10, 6))
sns.barplot(x="video_id", y="avg_latency", hue="window_size", data=df)
plt.title("Average Latency by Video and Window Size")
plt.xlabel("Video ID")
plt.ylabel("Average Latency (s)")
plt.xticks(rotation=45)
plt.legend(title="Window Size (s)")
plt.tight_layout()
plt.show()

# %% [markdown]
# ## P95 Latency Comparison by Window Size

# %%
plt.figure(figsize=(10, 6))
sns.barplot(x="video_id", y="p95_latency", hue="window_size", data=df)
plt.title("P95 Latency by Video and Window Size")
plt.xlabel("Video ID")
plt.ylabel("P95 Latency (s)")
plt.xticks(rotation=45)
plt.legend(title="Window Size (s)")
plt.tight_layout()
plt.show()
```
