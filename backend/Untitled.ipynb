```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pick top 10 videos by first appearance
top_videos = df['video_id'].unique()[:10]
df_top = df[df['video_id'].isin(top_videos)]

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Lineplot with markers, solid lines, and nicer palette
sns.lineplot(
    data=df_top,
    x="window_size",
    y="avg_latency",
    hue="video_id",
    style="video_id",   # different line styles for clarity
    markers=True,       # show markers
    dashes=False,       # solid lines
    linewidth=2.5,
    palette="tab10"     # palette for up to 10 distinct colors
)

# Titles and labels
plt.title("Average Latency per Video and Window Size (Top 10 Videos)", fontsize=16, weight='bold')
plt.xlabel("Window Size", fontsize=12)
plt.ylabel("Average Latency (seconds)", fontsize=12)

# Legend outside plot
plt.legend(title="Video ID", bbox_to_anchor=(1.05, 1), loc='upper left')

# Remove top and right spines
sns.despine(trim=True)

plt.tight_layout()
plt.show()
```
